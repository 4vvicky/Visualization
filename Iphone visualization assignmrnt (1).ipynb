# JOINTPLOT

```python
vig.jointplot('SalePrice','Mrp',p)
vi.show()
```

##### In this joinplot we have chosen a certain value and condition for the two variables "Saleprice" and "Mrp" where mrp less 40000 and Saleprice than  

```python
p=p[(p.SalePrice<50000)&(p.Mrp<80000)]
vig.jointplot('SalePrice','Mrp',p)
vi.show()
```

##### in this jointplot we have chosen a certain value and condition for the two variables "Saleprice" and "Mrp" where Mrp less RS 55000 and Saleprice less than 50000 in which there is a outlier plotted in 54000 in xaxis RS 42500 and RS 47500 Salesprice, in this condition there are RS 30000 Salesprice has minimum value the histogram for Salesprice plotted at the top of the joint plot and for Mrp it is plotted at te right side of the jointplot. for Mrp the histogram has maximummvalue present in the RS 52000 to RS 54000 and histogram of Saleprice shows us the RS 42500 to RS 47500 values are present in that data of the dataset.

```python
p=p[(p.SalePrice<50000)&(p.Mrp<80000)]
vig.jointplot('SalePrice','Mrp',p)
vig.jointplot(x='SalePrice',y='Mrp',data=p,kind='hex',color='r')
vi.show()
```

##### we have displayed the variables of Saleprice and Mrp with a certain condition and also using color and kind where in this we have given color as red and kind as hexagon by using hexagon when the hexagon becomes darker it shows us there are maximum values are present in that if it becomes lighter it shows us there are minimum value are there, in this the most the dark hexagon is plotted in the value of (RS 54000,Rs 45000) (Saleprice,Mrp) it also does displays the histograms of the respective variables.
